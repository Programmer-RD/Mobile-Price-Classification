## Modeling

```python
X = data.drop('price_range',axis=1)
y = data['price_range']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
def fit_and_score(models,X_train,X_test,y_train,y_test):
    model_scores = {}
    for name,model in models.items():
        model.fit(X_train,y_train)
        print(name)
        print(model.score(X_test,y_test))
        model_scores[name] = model.score(X_test,y_test)
    return model_scores
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
models = {
    'Random Forrest Classifier':RandomForestClassifier(),
    'Random Forrest Regressor':RandomForestRegressor(),
    'Logisitic Regressor':LogisticRegression(),
    'GaussianNB Classifier':GaussianNB(),
    'SGDClassifier Classifier':SGDClassifier(),
}
fit_and_score(models,X_train,X_test,y_train,y_test)
```

```python
models_scores_df = pd.DataFrame(fit_and_score(models,X_train,X_test,y_train,y_test),index=['Accuracy (%)'])
models_scores_df.T.plot.bar(figsize=(10,6))
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
```

```python
model = RandomForestRegressor()
data = data.sample(frac=1)
X = data.drop('price_range',axis=1)
y = data['price_range']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
model.fit(X_train,y_train)
```

```python
pickle.dump(model,open('model.pkl','wb'))
```

```python
grid = {
    "n_estimators":[100,500,1000,1500,2000,2500],
    "max_depth":[25,50],
    "max_features":["auto","sqrt",'log2'],
    "min_samples_split":[2,4],
    "min_samples_leaf":[2,4],
    'bootstrap':[False,True],
    'warm_start':[False,True]
}
model_rs = RandomizedSearchCV(RandomForestRegressor(),n_iter=25,cv=5,verbose=25,param_distributions=grid).fit(X_train,y_train)
print(model_rs.score(X_test,y_test))
print(model_rs.best_estimator_)
```

```python
pickle.dump(model_rs,open('./model_rs.pkl','wb'))
```

```python
grid = {
    "n_estimators":[100,500,1000,5000,10000],
    "max_depth":[25,50],
    "max_features":["auto","sqrt",'log2'],
    "min_samples_split":[2,4],
    "min_samples_leaf":[2,4],
    'bootstrap':[False,True],
    'warm_start':[False,True]
}
model_gs = GridSearchCV(RandomForestRegressor(),cv=round(2.5),verbose=25,param_grid=grid).fit(X_train,y_train)
print(model_gs.score(X_test,y_test))
print(model_gs.best_estimator_)
pickle.dump(model_gs,open('./model_grid.pkl','wb'))
```
